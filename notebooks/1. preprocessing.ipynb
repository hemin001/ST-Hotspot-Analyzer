```python
import pandas as pd

file_path = "Dataset/Raw/Crimes_2001_to_Present.csv" 

df = pd.read_csv(file_path)

df.info()
```

```python
df.head()
```

```python
summary = pd.DataFrame({
    "column": df.columns,
    "dtype": df.dtypes.astype(str),
    "missing_values": df.isna().sum().values
})

summary
```

```python
import pandas as pd

# df already loaded
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

use_cols = ["Date", "Primary Type", "Latitude", "Longitude", "Year"]
df = df[use_cols].copy()

# Drop rows with missing essential fields
df = df.dropna(subset=["Date", "Latitude", "Longitude"])

# Basic sanity
print(df.shape)
print(df["Year"].min(), df["Year"].max())
df.head()
```

```python
df.to_csv("Dataset/Processed/cleaned_crime.csv", index=False)
```
